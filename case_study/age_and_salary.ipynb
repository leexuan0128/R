# Case study: Age and Salary Analysis using R

```r
# Build Data Framework (Name Age and Salary)
Names  =  c("Mike  Hussey",  "Aaron  Finch",  "Brad  Hogg",  "Steve  Smith", "George  Bailey",  
            "Mitchell  Johnson",  "Shaun  Marsh",  "Glenn  Maxwel l",  "Pat  Cummins",  "Mitchell  Starc",  "David  Warner")

Ages = c(39,28,44,25,32,33,31,26,22,25,28)

IPLSals  =  c(310,662,103,828,672,1340,455,1240,207,1030,1140)

# Combine different attributes vectors
df = data.frame(Names, Ages, IPLSals)
# Show dataframe
df
```

```r
# Retrive data from dataframe
# The first row and all columns data
df[1, ]
```

```r
# All rows and the first column data
df[, 1]
```

```r
df[2:3] # column2 and 3
```

```r
df[2:3, ] # 2-3 row, all columns
```

## Determine Data Statistics

```r
# Average age? 3 methods
mean(df[, 2])
mean(df[, 'Ages'])
mean(df$Ages)
```

```r
var(df[, 2])
sd(df[, 2])
```

## Display Data in R

```r
# display data adding a linear regression
plot(df$Ages  ~  df$IPLSals,  main  =  "Cricket$") # plot Age
```

```r
plot(df$Ages  ~  df$IPLSals,  main  =  "Cricket$")
abline(lm(df$Ages  ~  df$IPLSals),  col  =  "red")	# Add a linear regression
```

## Determine the equation of the red line

```r
# use lm() to simulate the linear regression
fit  =  lm(df$Ages  ~  df$IPLSals) 
fit
# So the regression equation is: age = -0.006372 * salary + 34.899143
```

## lay down multiple plots

```r

sorted  =  sort(Ages)  #  we  can  see  some  other  data 
plot(Ages,  IPLSals,  xlab  =  "Ages",  col  =  "red"  ) 
par(new=TRUE)  #  adds  new  plot  to  old;  redraws  axes…. 
plot(sorted,  IPLSals,  xlab  =  "Ages",  col  =  "green"  )
```

```r
# or add points to a plot
plot(Ages,  IPLSals,  col  =  "blue"  ) 
points(sorted,  IPLSals,  col  =  "red"  )
```
